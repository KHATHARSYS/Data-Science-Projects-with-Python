```python
#Complejidad de Grado 1 (modelo lineal)
N_simul=100 #numero de modelos lineales
n=500
#intercepto=[]
#coef=[]
mu_x=0
sigma2_x=4
mu_y=0
sigma2_y=4
rho_2=0.04
```

```python
#Haremos 100 modelos lineales

y_out_test=np.zeros(150) #hacemos vector con 150ceros
x_out_test=np.zeros(150)
x_graf=np.zeros(60)
x_graf= np.reshape(x_graf, (1, 60))
y_graf= np.reshape(x_graf, (1, 60))


```

```python
y_out_test= np.reshape(x_out_test, (1, 150))
x_out_test= np.reshape(x_out_test, (1, 150))
y_fit_test= np.reshape(x_out_test, (1, 150))

y_fit_test
```

```python


#indices=#[str(i) for i in range(1,N_simul+1)]
#Generamos los modelos y lo guardamos como .sav

for k in range(1,N_simul):
        z_1=np.random.normal(0,1,n)
        z_2=np.random.normal(0,1,n)
        X=mu_x+sqrt(sigma2_x)*z_1
        y=mu_y+sqrt(sigma2_y)*(rho_2*z_1**5+sqrt(1-(rho_2)**2)*z_2)
        X=X[:, np.newaxis]
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
        
        pr = LinearRegression()
        quadratic = PolynomialFeatures(degree=1)
        X_quad = quadratic.fit_transform(X_train)
        pr.fit(X_quad, y_train)
        y_test_fit_2 = pr.predict(quadratic.fit_transform(X_test))
        X_fit_gr = np.arange(-6, 6, 0.2)[:, np.newaxis]
        y_test_graf = pr.predict(quadratic.fit_transform(X_fit_gr))
        X_fit_gr=X_fit_gr.flatten()
        y_test_graf=y_test_graf.flatten()
        yte_aux=y_test.flatten()
        yte_fit=y_test_fit_2.flatten()
        xte_aux=X_test.flatten()
        y_out_test=np.append(y_out_test,[yte_aux],axis= 0)
        x_out_test=np.append(x_out_test,[xte_aux],axis= 0)
        y_fit_test=np.append(y_fit_test,[yte_fit],axis= 0)
        x_graf=np.append(x_graf,[X_fit_gr],axis= 0)
        y_graf=np.append(y_graf,[y_test_graf],axis= 0)
fig, ax=subplots(figsize=(8,6))
scatter(X_train_o,y_train_o,s=65,color="gray",edgecolor="gray")
for j in range(1,15):
    plot(x_graf[j],y_graf[j],color="cadetblue")
ax.set_ylim([-10, 15])
xticks(fontsize=16)
yticks(fontsize=16)
xlabel('Muestra $x_{1},\ldots, x_{N}$',fontsize=16)
ylabel('$E[Y|X]=f(X)+\epsilon$, grado(f)=1',fontsize=16)
title("Ajustando un Modelo Lineal",fontsize=20)
savefig("_grado1.pdf")
```
